```python
data1 = pd.DataFrame({"key1":[1,2,3,4,5],
                      "key2":[4,5,6,7,8],
                      "key3":[3,4,5,6,7]
                     })
```

```python
data1
```

```python
data2 = pd.DataFrame({"key1":[12,24,43,4,5],
                      "key4":[4,5,6,7,8],
                      "key5":[3,4,5,16,7]
                     })
```

```python
data1
```

```python
data2
```

```python
# Merge operation
pd.merge(data1,data2)
# venn- diagram >> inner-join  (by default inner-merge works)
                # >> left- outer join, etc
```

```python
pd.merge(data1,data2,how= 'left')  
#left-merge >>take every cols from df1 and map to df2
```

```python
pd.merge(data1,data2,how= 'right')  
#right-merge >>take every cols from df2 and map to df1
```

```python
pd.merge(data1,data2,how= 'outer')  
#outer-merge >>take every cols from df1 and df2 > match shows otherwise NaN
```

```python
pd.merge(data1,data2,how= 'cross')  
#outer-merge >>matches every cols and maps with every rows of df1 and df2
```

```python
# Using 'ON' >> should specify which col to match
pd.merge(data1,data2,how= 'outer',on='key1')   #by default it maps to key1
```

# More operations on DataFrames:

# Difference b/w Merge and join
join is similar to merge but merge checks col data and map if matches

join works for indexes, where index is matching map with data

all inner, outer, left-outer can be performed in both join and merge

By default join >> uses left-join operation

```python
data1.join(data2)
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='cross')
```

```python
# Converting Euro Fare in INR > and create new row to store
# Using custom/ outside function:
# just like - map function of list
def euro_to_inr(x):
    return x*80

df['Fare_INR'] = df['Fare'].apply(euro_to_inr) #custom function
```

```python
# # Categorize Fare into a new column:
# 10 < 0 : cheap
# 10 - 20 : mid
# 20 < ++ : high

def categ_fare(x):
    if x < 10:
        return 'cheap'
    elif x >=10 and x <20:
        return 'mid'
    else:
        return 'high'
```

```python
df['Categorize_Fare'] = df['Fare'].apply(categ_fare)
```
